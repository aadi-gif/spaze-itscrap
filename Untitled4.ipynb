```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
from selenium.webdriver.chrome.options import Options

chrome_options = Options()
chrome_options.add_argument("--headless")

browser=webdriver.Chrome(options=chrome_options)
import time
browser.get("https://google.com")
browser.find_element(By.CLASS_NAME,"gLFyf").send_keys("spaze itech park companies")
time.sleep(3)
browser.find_element(By.CLASS_NAME,"gNO89b").click()
time.sleep(5)
browser.find_element(By.CLASS_NAME,"Z4Cazf.OSrXXb").click()
time.sleep(5)
try:
    get_lis = [j.text for j in browser.find_element(By.CLASS_NAME,"AaVjTc").find_elements(By.TAG_NAME,'a')]
    get_lis = [j for j in get_lis if j.isdigit()]
    get_last_page_no = int(get_lis[-1])
    print("page no",get_last_page_no)

    
    count=0
    company_lis= []
    for l in range(get_last_page_no-1):
        try:
            if count!=0:
                browser.find_element(By.XPATH,'//*[@id="pnnext"]/span[2]').click()
            get_val=browser.find_elements(By.CLASS_NAME,"VkpGBb")
            print("length of page",len(get_val))

            for j in range(len(get_val)):
                time.sleep(2)
                try:
                    get_val=browser.find_elements(By.CLASS_NAME,"VkpGBb")
                    get_val[j].click()
                    time.sleep(2)
                    get_name=browser.find_element(By.CLASS_NAME,"SPZz6b")

                    get_adrress=browser.find_element(By.CLASS_NAME,"LrzXr")
                    print("company name - ",get_name.text," Address - ",get_adrress.text)
                    company_lis.append((get_name.text,get_adrress.text))

                except:
                    continue
            count+=1
            print("\n\n\n")
        except:
            print(f"Page no {l} Break")
            continue
except Exception as e:
    print("Script Crash due to "+str(e))
df = pd.DataFrame(company_lis)
df.drop_duplicates(inplace=True)
df.to_csv("Spaze-It_Comapnies.csv",index=False)
```
